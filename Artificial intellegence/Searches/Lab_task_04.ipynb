```python
import networkx as nx
import queue
import pandas as pd

                                                        # FUNCTIONS

def breadthfs(graph,start,goal):
    myqueue = queue.Queue()
    myqueue.put((start,[start]))
    visited = set()
    
    while myqueue:
        node,path = myqueue.get()
        if node == goal:
            return path
        visited.add(node)
        
        for i in graph.neighbors(node):
            if i not in visited:
                myqueue.put((i , path + [i]))
    
    return []

def depthfs(graph,start,goal):
  
    stack.append((start,[start]))
    while stack:
        node,path = stack.pop()
        if node == goal:
            return path
        
        if node not in visited:
            visited.add(node)
            for neighbor in graph.neighbors(node):
                if neighbor is not visited:
                    stack.append((neighbor, path+[neighbor]))
    return []


def bestfunc(G,start,goal,heuristic):
    priorQueue = queue.PriorityQueue()
    priorQueue.put((heuristic[start],[start]))

    while not priorQueue.empty():
        heuristic_val, path = priorQueue.get()
        node = path[-1]

        if(node == goal):
            return path
        
        visited.add(node)

        for i in G.neighbors(node):
            if i not in visited:
                new_path = path + [i]                
                priorQueue.put((heuristic[i],new_path))
    
    return []

                                                        # MAIN FUNCTION

data = pd.read_csv("dummy_input.csv")
visited = set()
heuristic = {}
stack = []

source = list(data["source"])
destination = list(data["destination"])
weight = list(data["weight"])

G = nx.Graph()
edges = zip(source,destination,weight)
G.add_weighted_edges_from(edges)
# # Draw the graph
nx.draw(G, with_labels=True, node_color='skyblue', node_size=100, edge_color='k', linewidths=2, font_size=15)

start = input("Enter the start: ")
goal = input("Enter the goal: ")
start = start.upper()
goal = goal.upper()

for node, weight in zip(source, weight):
    heuristic[node] = weight

while True:
    print("Choose a search algorithm: ")
    print("1- Breadth-First_Search (BFS)")
    print("2- Depth-First_Search (DFS)")
    print("3- Best-First_Search (BFS)")
    print("4- Exit")
    choice = input("Enter -> ")

    if choice == '1':
        result = breadthfs(G,start,goal)
    elif choice == '2':
        result = depthfs(G,start,goal)
    elif choice == '3':
        result = bestfunc(G,start,goal,heuristic)
    else:
        print("Exiting Program")
        break
    print("\n Path from {} to {} is:  {} \n".format(start,goal,' -> '.join(result)))


```
