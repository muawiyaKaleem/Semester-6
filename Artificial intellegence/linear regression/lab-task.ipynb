```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(pd.read_csv("HousingData.csv"))
```

```python
df = df.fillna(round(df.mean()))
```

## **Finding and Removing the outliers**

```python
min_out = []
max_out = []
```

```python
for i in df.columns:
    df = df.sort_values(by=i,ascending=True)
    Q1 = 0.25 * (1 + len(df))
    Q2 = 0.5 * (1 + len(df))
    Q3 = 0.75 * (1 + len(df))
# Q1 = df.quantile(0.25)
# Q2 = df.quantile(0.5)
# Q3 = df.quantile(0.75)
    
    IQR = Q3 - Q1
    min_out.append(Q1 - (1.5 * IQR))
    max_out.append(Q3 + (1.5 * IQR))
```
